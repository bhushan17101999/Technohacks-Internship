# Create a histogram or bar chart to visualizethe distribution of data in a dataset

```python
#imported Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read the Data

df = pd.read_csv("Iris.csv")
df.head()
```

# Create the Visualization

```python
sns.histplot(data = df, x='SepalLengthCm', hue="Species", kde=True)
plt.title("Histogram of sepal Length by Species")
plt.show()
```

```python
#sepal-width
sns.histplot(data = df, x="SepalWidthCm", hue="Species", kde=True)
plt.title("Histogram of sepal Width by Species")
```

```python
#petal-lenght


sns.histplot(data = df, x="PetalLengthCm", hue="Species", kde=True)
plt.title("Histogram of Petal Lenght by Species")
```

```python
#Petal-Width


sns.histplot(data = df, x="PetalWidthCm", hue="Species", kde=True)
plt.title("Histogram of Petal Width by Species")
```
